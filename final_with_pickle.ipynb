```python
def convert(lst):
    it = iter(lst) 
    dct = dict(zip(it, it)) 
    return dct 
```

```python
def sort_convert(o):
    m=sorted(o.items())
    n=dict(m)
    return n
    
```

```python
def search_word():
    string_input=input("ENTER THE STRING:")
    part=string_input.split()
    total_words_c = sum(c.values())
    total_words_d = sum(d.values())
    total_words_e = sum(e.values())
    total_words_f = sum(f.values())
    total_words_g = sum(g.values())
    c1=[]
    d1=[]
    e1=[]
    f1=[]
    g1=[]
    
    
    if len(part)>=6:
        for key,value in c.items():
            for i in range (0,5):
                if(part[-1]==key[i] and part[-2]==key[i-1] and part[-3]==key[i-2] and part[-4]==key[i-3] and part[-5]==key[i-4] ):
                    probability_c= value/ total_words_c
                    c1.append(probability_c)
                    c1.append(key[i+1])
                        
         #          print("{}    {}".format(key[i+1],probability_c))   #print individual probability
        
        c_dct=convert(c1)
        s=sort_convert(c_dct)
        print("\n SUGGESTED WORD    PROBABILITY")
        print("---------------- --------------")
        for x in list(s)[len(list(s)):len(list(s))-4:-1]:
            print ("   {}          {} ".format(s[x],x ))
        
    
    
    elif len(part)>=5:
        for key,value in d.items():
            for i in range (0,4):
                if(part[-1]==key[i] and part[-2]==key[i-1] and part[-3]==key[i-2] and part[-4]==key[i-3]):
                    probability_d= value/ total_words_d
                    d1.append(probability_d)
                    d1.append(key[i+1])
                        
                   # print("{}    {}".format(key[i+1],probability_d))
        d_dct=convert(d1)
        s1=sort_convert(d_dct)
        print("\n SUGGESTED WORD    PROBABILITY")
        print("---------------- --------------")
        for x in list(s1)[len(list(s1)):len(list(s1))-4:-1]:
            print ("   {}          {} ".format(s1[x],x ))
        
 
    
    elif len(part)>=4:
        for key,value in e.items():
            for i in range (0,3):
                if(part[-1]==key[i] and part[-2]==key[i-1] and part[-3]==key[i-2] ):
                    probability_e= value/ total_words_e
                    e1.append(probability_e)
                    e1.append(key[i+1])
                        
                    #print("{}    {}".format(key[i+1],probability_e))
        e_dct=convert(e1)
        s2=sort_convert(e_dct)
        print("\n SUGGESTED WORD    PROBABILITY")
        print("---------------- --------------")
        for x in list(s2)[len(list(s2)):len(list(s2))-4:-1]:
            print ("   {}          {} ".format(s2[x],x ))
       
    
    elif len(part)>=3:
        for key,value in f.items():
            for i in range (0,2):
                if(part[-1]==key[i] and part[-2]==key[i-1] ):
                    probability_f= value/ total_words_f
                    f1.append(probability_f)
                    f1.append(key[i+1])
                        
                  #  print("{}    {}".format(key[i+1],probability_f))
        f_dct=convert(f1)
        s3=sort_convert(f_dct)
        print("\n SUGGESTED WORD    PROBABILITY")
        print("---------------- --------------")
        for x in list(s3)[len(list(s3)):len(list(s3))-4:-1]:
            print ("   {}          {} ".format(s3[x],x ))
                    
      
    elif len(part)>=2:
        for key,value in g.items():
            for i in range (0,1):
                if(part[-1]==key[i] ):
                    probability_g= value/ total_words_g
                    g1.append(probability_g)
                    g1.append(key[i+1])
                        
                    #print("{}  {}".format(key[i+1],probability_g))
        g_dct=convert(g1)
        s4=sort_convert(g_dct)
        print("\n SUGGESTED WORD    PROBABILITY")
        print("---------------- --------------")
        for x in list(s4)[len(list(s4)):len(list(s4))-4:-1]:
            print ("   {}          {} ".format(s4[x],x ))
  


    
   
```
